```python
import os
import glob
from PIL import Image
import matplotlib.pyplot as plt

# 1. ĐỊNH NGHĨA ĐƯỜNG DẪN
# ---
# Đường dẫn đến thư mục "ảnh gốc" (Ground Truth)
gt_dir = "data/DIS5K/DIS-TE4/gt"
# Đường dẫn đến kết quả dự đoán của bạn (chọn 1 trong 2)
pred_dir = "output/output-macro/DIS-TE4"   # <-- Kết quả từ Bước 1
# pred_dir = "output/output-micro/DIS-TE4"  # <-- Kết quả từ Bước 2
# ---

# 2. Lấy danh sách tất cả các file dự đoán
# (Chúng ta sẽ chỉ xem 5 ảnh đầu tiên cho nhanh)
pred_paths = sorted(glob.glob(os.path.join(pred_dir, "*.png")))[0:10]

print(f"--- Hiển thị 10 ảnh so sánh từ: {pred_dir} ---")

# 3. Lặp qua từng file dự đoán
for pred_path in pred_paths:
    # Lấy tên file (ví dụ: '1#Accessories...o.png')
    file_name = os.path.basename(pred_path)

    # Tạo đường dẫn đến file "ảnh gốc" tương ứng
    gt_path = os.path.join(gt_dir, file_name)

    # 4. Kiểm tra xem file gốc có tồn tại không
    if os.path.exists(gt_path):
        # 5. Tải 2 ảnh lên
        gt_image = Image.open(gt_path).convert("L")  # 'L' = grayscale
        pred_image = Image.open(pred_path).convert("L")

        # 6. Vẽ 2 ảnh cạnh nhau
        fig, ax = plt.subplots(1, 2, figsize=(12, 6))

        ax[0].imshow(gt_image, cmap='gray')
        ax[0].set_title(f"Ảnh Gốc (Ground Truth)\n{file_name}")
        ax[0].axis('off')

        ax[1].imshow(pred_image, cmap='gray')
        ax[1].set_title(f"Ảnh Dự Đoán (Prediction)\n{file_name}")
        ax[1].axis('off')

        plt.show() # Hiển thị

    else:
        print(f"Không tìm thấy file Ground Truth cho: {file_name}")
```

```python
import cv2
import numpy as np
import os
import glob
from tqdm import tqdm

# --- CẤU HÌNH (Bạn có thể sửa dòng PRED_DIR) ---

# Đường dẫn đến thư mục "ảnh gốc" (Ground Truth)
GT_DIR = "data/DIS5K/DIS-TE4/gt"

# Đường dẫn đến kết quả dự đoán của bạn (CHỌN 1 TRONG 2)
# PRED_DIR = "output/output-macro/DIS-TE4"   # <-- Kết quả từ Bước 1
PRED_DIR = "output/output-macro/DIS-TE4"  # <-- Kết quả tinh chỉnh (Bước 2)

# --- Hết cấu hình ---

def calculate_metrics(pred_path, gt_path):
    """Hàm đọc 2 ảnh và tính MAE, IoU."""
    try:
        # Đọc ảnh, chuyển về ảnh xám (0-255)
        pred_img = cv2.imread(pred_path, cv2.IMREAD_GRAYSCALE)
        gt_img = cv2.imread(gt_path, cv2.IMREAD_GRAYSCALE)

        # Đảm bảo 2 ảnh có cùng kích thước
        if pred_img.shape != gt_img.shape:
            pred_img = cv2.resize(pred_img, (gt_img.shape[1], gt_img.shape[0]))

        # 1. Chuẩn hóa (Normalize) ảnh về 0-1
        # Đảm bảo GT là 0 hoặc 1
        gt_norm = (gt_img > 127).astype(np.float32)
        # Chuẩn hóa ảnh dự đoán (thường là ảnh xám)
        pred_norm = (pred_img / 255.0).astype(np.float32)

        # 2. Tính MAE (Mean Absolute Error)
        mae = np.mean(np.abs(pred_norm - gt_norm))

        # 3. Tính IoU (Intersection over Union)
        # Chuyển dự đoán thành ảnh nhị phân (0 hoặc 1)
        pred_bin = (pred_norm > 0.5).astype(np.float32)
        gt_bin = gt_norm # Đã là 0 hoặc 1

        intersection = np.sum(pred_bin * gt_bin)
        union = np.sum(pred_bin) + np.sum(gt_bin) - intersection

        # Thêm epsilon (1e-6) để tránh chia cho 0
        iou = (intersection + 1e-6) / (union + 1e-6)

        return mae, iou
    except Exception as e:
        print(f"Lỗi xử lý file {os.path.basename(pred_path)}: {e}")
        return None, None

# --- Bắt đầu chạy ---
pred_paths = sorted(glob.glob(os.path.join(PRED_DIR, "*.png")))

if not pred_paths:
    print(f"LỖI: Không tìm thấy file dự đoán nào trong: {PRED_DIR}")
else:
    print(f"Bắt đầu tính score cho {len(pred_paths)} ảnh trong {PRED_DIR}...")

    all_maes = []
    all_ious = []

    # Lặp (for) qua từng file dự đoán
    for pred_path in tqdm(pred_paths):
        file_name = os.path.basename(pred_path)
        gt_path = os.path.join(GT_DIR, file_name)

        if os.path.exists(gt_path):
            mae, iou = calculate_metrics(pred_path, gt_path)
            if mae is not None:
                all_maes.append(mae)
                all_ious.append(iou)
        else:
            # Bỏ qua nếu file đó không có trong bộ Ground Truth
            pass

    # 4. In kết quả cuối cùng
    if all_maes:
        avg_mae = np.mean(all_maes)
        avg_iou = np.mean(all_ious)
        print("\n--- KẾT QUẢ CUỐI CÙNG ---")
        print(f"Tổng số ảnh đã xử lý: {len(all_maes)}")
        print(f"MAE trung bình (Càng thấp càng tốt): {avg_mae:.4f}")
        print(f"IoU trung bình (Càng cao càng tốt): {avg_iou:.4f}")
    else:
        print("Không có ảnh nào được xử lý.")
```
